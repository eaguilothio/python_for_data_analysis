## 🧠 Teoría rápida — Variables y tipos de datos

Una **variable** es un nombre que apunta a un valor en memoria. Python infiere el tipo automáticamente.

```python
cantidad  = 3        # int   → número entero
precio    = 29.99    # float → número con decimales
producto  = "Camiseta"  # str   → texto
en_stock  = True     # bool  → True o False
```

**Imprimir con f-string:** es una forma de insertar variables dentro de un texto usando {}.
La f antes de las comillas indica que Python debe reemplazar las variables por su valor.
```python
print(f"{producto} cuesta {precio} €")
```

> 💡 `type(variable)` te dice qué tipo de dato es.

````python
# Datos de un producto en una tienda online
producto     = "Zapatillas running"
marca        = "Nike"
precio       = 89.99
stock        = 42
disponible   = True

print(f"{producto} de {marca} — {precio} € — disponible: {disponible}")
````

````python
# ¿De qué tipo es cada dato?
print(type(producto))
print(type(precio))
print(type(stock))
print(type(disponible))
````

## 🧠 Teoría rápida — Casting (conversión de tipos)

Al leer datos de un archivo es habitual que los números lleguen como texto. El casting permite convertirlos para operar.

```python
int("5")       # "5"    → 5
float("9.99")  # "9.99" → 9.99
str(99)        # 99     → "99"
```

> ⚠️ `"29.99" + 10` da error. Necesitas convertir primero: `float("29.99") + 10`

````python
# Estos valores llegaron como texto desde un archivo — ¿cómo los convertimos?
precio_str   = "89.99"
cantidad_str = "3"

precio   = float(precio_str)
cantidad = int(cantidad_str)

print(type(precio))
print(type(cantidad))
````

## 🧠 Teoría rápida — Operaciones matemáticas

| Operador | Significado      | Ejemplo  | Resultado |
|----------|------------------|----------|-----------|
| `+`      | suma             | `8 + 2`  | `10`      |
| `-`      | resta            | `8 - 2`  | `6`       |
| `*`      | multiplicación   | `8 * 2`  | `16`      |
| `/`      | división (float) | `8 / 2`  | `4.0`     |
| `//`     | división entera  | `9 // 2` | `4`       |
| `%`      | resto (módulo)   | `9 % 2`  | `1`       |
| `**`     | potencia         | `3 ** 2` | `9`       |

````python
# ¿Cuánto cuesta el pedido en total?
precio   = 89.99
cantidad = 3

total = round(precio * cantidad, 2)
print(f"Total del pedido: {total} €")
````

````python
# ¿Cuánto gana la tienda por cada unidad vendida?
precio_venta = 89.99
coste        = 45.00

beneficio_unitario = round(precio_venta - coste, 2)
print(f"Beneficio por unidad: {beneficio_unitario} €")
````

````python
# ¿Cuántos pedidos completos se pueden hacer con el stock actual?
stock_disponible = 50
unidades_por_pedido = 3

pedidos_posibles = stock_disponible // unidades_por_pedido
unidades_sobrantes = stock_disponible % unidades_por_pedido
print(f"Pedidos posibles: {pedidos_posibles} — Unidades sobrantes: {unidades_sobrantes}")
````

## 🧠 Teoría rápida — Comparaciones y lógica booleana

Las comparaciones devuelven `True` o `False` y son la base de los filtros en pandas.

```python
df[df["precio"] > 50]                                    # productos con precio > 50
df[(df["stock"] > 0) & (df["disponible"] == True)]      # en stock y disponibles
```

| Operador | Significado |
|----------|-------------|
| `>` `<` `>=` `<=` | mayor, menor, mayor o igual, menor o igual |
| `==` | igual |
| `!=` | distinto |
| `and` | las dos condiciones deben ser verdaderas |
| `or`  | basta con que una sea verdadera |

````python
# ¿El precio está por debajo de 100 €?
precio = 89.99
print(precio < 100)
````

````python
# ¿Hay suficiente stock para un pedido de 5 unidades?
stock           = 42
unidades_pedido = 5
print(stock >= unidades_pedido)
````

````python
# ¿El producto está disponible Y tiene stock?
disponible = True
stock      = 42
print(disponible and stock > 0)
````

````python
# ¿El precio es menor de 50 € O hay más de 100 unidades en stock?
precio = 89.99
stock  = 120
print(precio < 50 or stock > 100)
````

## 🧠 Teoría rápida — Métodos de strings

Al analizar datos pasarás mucho tiempo limpiando texto: nombres con mayúsculas inconsistentes, espacios extra, caracteres a reemplazar.

```python
texto.upper()            # TODO EN MAYÚSCULAS
texto.lower()            # todo en minúsculas
texto.strip()            # elimina espacios al principio y al final
texto.replace("a", "b") # reemplaza caracteres
texto.title()            # Primera Letra De Cada Palabra En Mayúscula
```

````python
# El nombre del producto llegó con espacios extra — ¿cómo lo limpiamos?
producto_raw = "  zapatillas running  "

producto_limpio = producto_raw.strip().title()
print(producto_limpio)
````

````python
# La categoría está en mayúsculas — ¿cómo la estandarizamos?
categoria_raw = "DEPORTES Y FITNESS"

categoria_limpia = categoria_raw.lower().title()
print(categoria_limpia)
````

````python
# El código de producto tiene guiones — ¿cómo los eliminamos?
codigo_raw = "NK-2024-RUN"

codigo_limpio = codigo_raw.replace("-", "")
print(codigo_limpio)
````
