## 🧠 Teoría rápida — Variables y tipos de datos

Una **variable** es un nombre que apunta a un valor en memoria. Python infiere el tipo automáticamente.

```python
edad     = 25       # int   → número entero
salario  = 2850.50  # float → número con decimales
nombre   = "Ana"    # str   → texto
activo   = True     # bool  → True o False
```

**Imprimir con f-string:**
```python
print(f"{nombre} cobra {salario} € al mes")
```

> 💡 `type(variable)` te dice qué tipo de dato es.

````python
# ¿Qué información básica tenemos de un empleado?
nombre          = "Laura García"
departamento    = "Marketing"
salario_mensual = 2850.50
años_empresa    = 4
activo          = True

print(f"{nombre} trabaja en {departamento} desde hace {años_empresa} años.")
````

````python
# ¿De qué tipo es cada dato?
print(type(nombre))
print(type(salario_mensual))
print(type(años_empresa))
print(type(activo))
````

## 🧠 Teoría rápida — Casting (conversión de tipos)

En análisis de datos es habitual encontrar números guardados como texto al leer un CSV. El casting permite convertirlos para operar.

```python
int("5")       # "5"    → 5
float("9.99")  # "9.99" → 9.99
str(99)        # 99     → "99"
```

> ⚠️ `"2850" + 500` da error. Necesitas convertir primero: `int("2850") + 500`

````python
# Estos valores llegaron como strings desde un CSV — ¿cómo los convertimos?
precio_str   = "149.99"
cantidad_str = "30"

precio   = float(precio_str)
cantidad = int(cantidad_str)

print(type(precio))
print(type(cantidad))
````

## 🧠 Teoría rápida — Operaciones matemáticas

| Operador | Significado      | Ejemplo  | Resultado |
|----------|------------------|----------|-----------|
| `+`      | suma             | `8 + 2`  | `10`      |
| `-`      | resta            | `8 - 2`  | `6`       |
| `*`      | multiplicación   | `8 * 2`  | `16`      |
| `/`      | división (float) | `8 / 2`  | `4.0`     |
| `//`     | división entera  | `9 // 2` | `4`       |
| `%`      | resto (módulo)   | `9 % 2`  | `1`       |
| `**`     | potencia         | `3 ** 2` | `9`       |

````python
# ¿Cuánto vale el inventario en total?
precio_unitario = 149.99
unidades        = 30

valor_inventario = precio_unitario * unidades
print(f"Valor total del inventario: {valor_inventario} €")
````

````python
# ¿Cuál es el precio final tras aplicar un descuento del 15%?
precio     = 149.99
descuento  = 0.15

precio_final = round(precio * (1 - descuento), 2)
print(f"Precio con descuento: {precio_final} €")
````

````python
# ¿Cuál es el precio con IVA incluido?
precio = 149.99
iva    = 0.21

precio_con_iva = round(precio + (precio * iva), 2)
print(f"Precio con IVA: {precio_con_iva} €")
````

````python
# ¿Cuál es el margen de beneficio?
precio_venta = 149.99
coste        = 85.00

margen = round((precio_venta - coste) / precio_venta * 100, 1)
print(f"Margen de beneficio: {margen}%")
````

````python
# ¿Cuánto cobra un empleado al año?
salario_mensual = 2850.50

salario_anual = salario_mensual * 12
print(f"Salario anual: {salario_anual} €")
````

## 🧠 Teoría rápida — Comparaciones y lógica booleana

Las comparaciones devuelven `True` o `False` y son la base de los filtros en pandas.

```python
df[df["salario"] > 2000]                              # filas donde salario > 2000
df[(df["ventas"] > 3000) & (df["activo"] == True)]   # condiciones combinadas
```

| Operador | Significado |
|----------|-------------|
| `>` `<` `>=` `<=` | mayor, menor, mayor o igual, menor o igual |
| `==` | igual |
| `!=` | distinto |
| `and` | las dos condiciones deben ser verdaderas |
| `or`  | basta con que una sea verdadera |

````python
# ¿Se ha superado el objetivo de ventas del mes?
objetivo_ventas = 3000
ventas_mes      = 3450

print(ventas_mes > objetivo_ventas)
````

````python
# ¿El salario del empleado está por encima del mínimo establecido?
salario_minimo  = 2500
salario_mensual = 2850.50

print(salario_mensual >= salario_minimo)
````

````python
# ¿El empleado lleva más de 3 años en la empresa Y sigue activo?
años_empresa = 4
activo       = True

print(años_empresa > 3 and activo)
````

````python
# ¿Las ventas superan el objetivo O el ticket medio supera los 100 €?
ventas_mes     = 3450
objetivo_ventas = 3000
ticket_medio   = 85

print(ventas_mes > objetivo_ventas or ticket_medio > 100)
````

## 🧠 Teoría rápida — Métodos de strings

En análisis de datos pasarás mucho tiempo limpiando texto: nombres con mayúsculas inconsistentes, espacios extra, caracteres a reemplazar.

```python
texto.upper()           # TODO EN MAYÚSCULAS
texto.lower()           # todo en minúsculas
texto.strip()           # elimina espacios al principio y al final
texto.replace("a","b") # reemplaza caracteres
texto.title()           # Primera Letra De Cada Palabra En Mayúscula
```

````python
# El nombre llegó con espacios extra — ¿cómo lo limpiamos?
nombre_raw = "  laura garcía  "

nombre_limpio = nombre_raw.strip().title()
print(nombre_limpio)
````

````python
# El departamento está en mayúsculas — ¿cómo lo estandarizamos?
departamento_raw = "MARKETING"

departamento_limpio = departamento_raw.lower().capitalize()
print(departamento_limpio)
````

````python
# El teléfono tiene guiones — ¿cómo los eliminamos?
telefono_raw = "612-345-678"

telefono_limpio = telefono_raw.replace("-", "")
print(telefono_limpio)
````
