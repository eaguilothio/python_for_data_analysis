## 🧠 Teoría rápida — Listas

Una **lista** guarda varios valores en orden bajo un mismo nombre.

```python
precios = [29.99, 49.99, 15.99, 89.99]
```

**Acceder por índice** (empieza en 0):
```python
precios[0]   # 29.99 → primero
precios[-1]  # 89.99 → último
```

**Métodos más usados:**
```python
len(lista)          # número de elementos
max(lista)          # valor máximo
min(lista)          # valor mínimo
sum(lista)          # suma de todos los elementos
lista.sort()        # ordena de menor a mayor (modifica el original)
lista.copy()        # crea una copia independiente
lista.count(x)      # cuántas veces aparece x
lista.index(x)      # posición de x
lista.append(x)     # añade x al final
x in lista          # True/False → ¿está x?
```

> 💡 En pandas, una columna es esencialmente una lista. Todo lo que practiques aquí tiene traducción directa.

````python
# Precios de los productos del carrito de un cliente (en euros)
precios = [29.99, 49.99, 9.99, 89.99, 15.99, 34.99]
````

````python
# ¿Cuántos productos hay en el carrito?
print(len(precios))
````

````python
# ¿Cuál es el producto más caro?
print(max(precios))
````

````python
# ¿Cuál es el producto más barato?
print(min(precios))
````

````python
# ¿Cuánto suma el carrito en total?
print(round(sum(precios), 2))
````

````python
# ¿Cuál es el precio medio por producto?
total    = sum(precios)
cantidad = len(precios)
media    = round(total / cantidad, 2)

print(f"Precio medio por producto: {media} €")
````

````python
# ¿Cuál es el precio del primer producto añadido al carrito?
print(precios[0])
````

````python
# ¿Cuál es el precio del último producto añadido?
print(precios[-1])
````

````python
# ¿Qué producto es el más caro y en qué posición está?
productos = ["Camiseta", "Pantalón", "Calcetines", "Zapatillas", "Gorra", "Sudadera"]

precio_mas_caro = max(precios)
posicion        = precios.index(precio_mas_caro)

print(f"El producto más caro es '{productos[posicion]}' con {precio_mas_caro} €")
````

````python
# ¿Hay algún producto que cueste exactamente 20 €?
print(20 in precios)
````

````python
# ¿Cuántos productos cuestan 9.99 €?
print(precios.count(9.99))
````

````python
# Ordena los precios de mayor a menor para ver el ranking.
# ⚠️ .sort() modifica la lista original — trabaja sobre una copia. 
# Ordena la lista de mayor a menor (reverse=True = orden descendente)
precios_ranking = precios.copy()
precios_ranking.sort(reverse=True)
print(precios_ranking)
````

````python
# El cliente añade un producto más al carrito: unos auriculares a 59.99 €.
precios.append(59.99)
print(f"Productos en el carrito: {len(precios)}")
print(f"Último precio añadido:   {precios[-1]} €")
````

## 📋 Resumen de lo aprendido

| Operación | Código |
