## 🧠 Teoría rápida — Listas

Una **lista** guarda varios valores en orden bajo un mismo nombre.

```python
ventas = [1200, 980, 1450, 870]
```

**Acceder por índice** (empieza en 0):
```python
ventas[0]   # 1200 → primero
ventas[-1]  # 870  → último
```

**Métodos más usados:**
```python
len(lista)          # número de elementos
max(lista)          # valor máximo
min(lista)          # valor mínimo
sum(lista)          # suma de todos los elementos
lista.sort()        # ordena de menor a mayor (modifica el original)
lista.copy()        # crea una copia independiente
lista.count(x)      # cuántas veces aparece x
lista.index(x)      # posición de x
lista.append(x)     # añade x al final
x in lista          # True/False → ¿está x?
```

> 💡 En pandas, una columna es esencialmente una lista. Todo lo que practiques aquí tiene traducción directa.

````python
# Ventas mensuales del año (en miles de euros)
ventas = [120, 98, 145, 87, 160, 134, 112, 99, 178, 155, 190, 210]
````

````python
# ¿Cuántos meses hay registrados?
print(len(ventas))
````

````python
# ¿Cuál fue la venta más alta del año?
print(max(ventas))
````

````python
# ¿Cuál fue la venta más baja del año?
print(min(ventas))
````

````python
# ¿Cuánto se vendió en total durante el año?
print(sum(ventas))
````

````python
# ¿Cuál fue la media mensual de ventas?
total_ventas  = sum(ventas)
numero_meses  = len(ventas)
media_mensual = round(total_ventas / numero_meses, 1)

print(f"Media mensual: {media_mensual} k€")
````

````python
# ¿Cuánto se vendió en enero (primer mes)?
print(ventas[0])
````

````python
# ¿Cuánto se vendió en diciembre (último mes)?
print(ventas[-1])
````

````python
# ¿En qué mes se registró la venta más alta?
meses  = ["enero", "febrero", "marzo", "abril", "mayo", "junio",
          "julio", "agosto", "septiembre", "octubre", "noviembre", "diciembre"]

venta_mas_alta = max(ventas)
posicion       = ventas.index(venta_mas_alta)

print(f"La venta más alta fue en {meses[posicion]}")
````

````python
# ¿Hubo algún mes con ventas de exactamente 100 k€?
print(100 in ventas)
````

````python
# ¿Cuántas veces se alcanzaron ventas de 98 k€?
print(ventas.count(98))
````

````python
# Ordena las ventas de mayor a menor para ver el ranking.
# ⚠️ .sort() modifica la lista original — trabaja sobre una copia.
ventas_ranking = ventas.copy()
ventas_ranking.sort(reverse=True)
print(ventas_ranking)
````

````python
# Acaba de cerrar enero del año siguiente con 225 k€.
# Añádelo a la lista original.
ventas.append(225)
print(f"Total de registros: {len(ventas)}")
print(f"Último valor:       {ventas[-1]} k€")
````
