## 🧠 Teoría rápida — Diccionarios

Un **diccionario** guarda pares `clave: valor`. La clave es única.

```python
producto = {
    "nombre":    "Zapatillas running",
    "precio":    89.99,
    "categoria": "Deportes"
}
```

**Acceder, añadir y modificar:**
```python
producto["nombre"]           # acceder → "Zapatillas running"
producto["marca"] = "Nike"   # añadir nueva clave
producto["precio"] = 79.99   # modificar valor existente
```

**Métodos más usados:**
```python
d.keys()                   # lista de claves
d.values()                 # lista de valores
d.items()                  # lista de pares (clave, valor)
d.get("clave", "defecto")  # accede sin error si no existe
d.update({"k": "v"})       # añade o actualiza entradas
```

> 💡 En pandas usarás diccionarios para construir DataFrames (`pd.DataFrame(dict)`), renombrar columnas (`.rename(columns={...})`) y mapear valores (`.map(dict)`).

````python
# Ficha de un producto en la tienda
producto = {
    "nombre":    "Zapatillas running",
    "marca":     "Nike",
    "precio":    89.99,
    "stock":     42,
    "disponible": True
}
````

````python
# ¿Cuánto cuesta el producto?
print(producto["precio"])
````

````python
# ¿Cuántas unidades quedan en stock?
print(producto["stock"])
````

````python
# La tienda hace una rebaja — el precio baja a 74.99 €.
producto["precio"] = 74.99
print(producto["precio"])
````

````python
# El producto cambia de categoría y se añade la talla disponible.
# .update() permite modificar varios campos a la vez en una sola línea.
producto.update({"categoria": "Running", "talla": "42"})
print(producto)
````

````python
# ¿Qué campos tiene la ficha de este producto?
print(list(producto.keys()))
````

````python
# ¿Cuáles son todos los valores de la ficha?
print(list(producto.values()))
````

````python
# ¿Cuáles son todos los pares clave-valor de la ficha?
print(list(producto.items()))
````

````python
# ¿Tiene registrada la valoración media el producto?
# Usamos .get() para evitar un error si la clave no existe. Si no existe → devuelve "sin valoraciones aún".
print(producto.get("valoracion", "sin valoraciones aún"))
````

````python
# Lista de diccionarios — estructura habitual antes de crear un DataFrame
carrito = [
    {"nombre": "Zapatillas running", "precio": 74.99, "cantidad": 1},
    {"nombre": "Calcetines deportivos", "precio": 9.99,  "cantidad": 3},
    {"nombre": "Botella de agua",      "precio": 14.99, "cantidad": 2},
]

# ¿Cuántos productos distintos hay en el carrito?
print(len(carrito))
````

````python
# ¿Cuánto cuesta el segundo producto del carrito?
print(carrito[1]["precio"])
````

## 📋 Resumen de lo aprendido

| Lista de claves | `list(d.keys())` |
| Lista de valores | `list(d.values())` |
| Lista de pares | `list(d.items())` |
