## 🧠 Teoría rápida — Diccionarios

Un **diccionario** guarda pares `clave: valor`. La clave es única.

```python
empleado = {
    "nombre":     "Laura García",
    "salario":    2850.50,
    "departamento": "Marketing"
}
```

**Acceder, añadir y modificar:**
```python
empleado["nombre"]          # acceder → "Laura García"
empleado["ciudad"] = "BCN"  # añadir nueva clave
empleado["salario"] = 3000  # modificar valor existente
```

**Métodos más usados:**
```python
d.keys()                   # lista de claves
d.values()                 # lista de valores
d.items()                  # lista de pares (clave, valor)
d.get("clave", "defecto")  # accede sin error si no existe
d.update({"k": "v"})       # añade o actualiza entradas
```

> 💡 En pandas usarás diccionarios para construir DataFrames (`pd.DataFrame(dict)`), renombrar columnas (`.rename(columns={...})`) y mapear valores (`.map(dict)`).

````python
# Registro de un empleado
empleado = {
    "nombre":        "Laura García",
    "departamento":  "Marketing",
    "salario":       2850.50,
    "años_empresa":  4,
    "activo":        True
}
````

````python
# ¿Cuánto cobra el empleado?
print(empleado["salario"])
````

````python
# ¿En qué departamento trabaja?
print(empleado["departamento"])
````

````python
# El empleado ha recibido una subida de salario a 3100 €.
empleado["salario"] = 3100
print(empleado["salario"])
````

````python
# El empleado se traslada a Barcelona y cambia de departamento.
# .update() permite modificar varios campos a la vez en una sola línea.
empleado.update({"ciudad": "Barcelona", "departamento": "Ventas"})
print(empleado)
````

````python
# ¿Qué campos tiene el registro de este empleado?
print(list(empleado.keys()))
````

````python
# ¿Cuáles son todos los valores del registro?
print(list(empleado.values()))
````

````python
# ¿Cuáles son todos los pares clave-valor del registro?
print(list(empleado.items()))
````

````python
# ¿Tiene registrado el email el empleado?
# Usamos .get() para evitar un error si la clave no existe.
print(empleado.get("email", "sin email registrado"))
````

````python
# Lista de diccionarios — estructura habitual antes de crear un DataFrame
empleados = [
    {"nombre": "Laura García",  "departamento": "Marketing", "salario": 3100},
    {"nombre": "Carlos Ruiz",   "departamento": "Ventas",    "salario": 2700},
    {"nombre": "Marta Sanz",    "departamento": "Marketing", "salario": 3400},
    {"nombre": "Iván López",    "departamento": "Ventas",    "salario": 2500},
]

# ¿Cuántos empleados hay en el registro?
print(len(empleados))
````

````python
# ¿Cuál es el salario del tercer empleado?
print(empleados[2]["salario"])
````

## 📋 Resumen de lo aprendido

| Operación | Código |
|---|---|
| Crear diccionario | `d = {"clave": "valor"}` |

| Modificar valor | `d["clave"] = nuevo_valor` |
| Añadir / actualizar | `d.update({"k": "v"})` |
| Lista de claves | `list(d.keys())` |
| Lista de valores | `list(d.values())` |
| Lista de pares | `list(d.items())` |
