## 🧠 Teoría rápida — if / elif / else

Permite que el programa elija qué hacer según una condición.

```python
stock = 5

if stock > 0:
    print("En stock")
else:
    print("Agotado")
```

**Reglas importantes:**
- Solo se ejecuta el **primer bloque** cuya condición sea verdadera.
- `else` es el caso por defecto si ninguna condición fue True.
- La **indentación** (4 espacios) es obligatoria.

> 💡 En análisis de datos usarás `if/elif/else` dentro de funciones aplicadas con `.apply()` para clasificar o etiquetar valores en una columna.

````python
# ¿Hay suficiente stock para completar el pedido?
stock_disponible = 45
unidades_pedido  = 60

if stock_disponible >= unidades_pedido:
    print("Stock suficiente — pedido confirmado")
else:
    print("Stock insuficiente — pedido rechazado")
````

````python
# ¿El total del carrito supera el mínimo para envío gratuito (40 €)?
total_carrito    = 53.97
minimo_envio_gratis = 40

if total_carrito >= minimo_envio_gratis:
    print("¡Envío gratuito!")
else:
    faltan = round(minimo_envio_gratis - total_carrito, 2)
    print(f"Te faltan {faltan} € para el envío gratuito")
````

````python
# ¿El cliente tiene cuenta de usuario Y está verificado?
tiene_cuenta = True
verificado   = True

if tiene_cuenta and verificado:
    print("Acceso permitido — bienvenido")
else:
    print("Acceso denegado")
````

## 🧠 Teoría rápida — elif (múltiples categorías)

Cuando hay más de dos opciones posibles, usamos `elif`:

```python
if total >= 100:
    descuento = "15%"
elif total >= 50:
    descuento = "10%"
else:
    descuento = "Sin descuento"
```

> 💡 Python evalúa las condiciones de arriba a abajo y ejecuta solo la primera que sea True.

````python
# ¿Qué descuento le corresponde al cliente según el total de su pedido?
total_pedido = 75

if total_pedido >= 100:
    descuento = "15%"
elif total_pedido >= 50:
    descuento = "10%"
elif total_pedido >= 25:
    descuento = "5%"
else:
    descuento = "Sin descuento"

print(f"Total: {total_pedido} € → Descuento: {descuento}")
````

````python
# ¿En qué estado está el pedido según los días transcurridos?
dias_desde_pedido = 3

if dias_desde_pedido == 0:
    estado = "Recién realizado"
elif dias_desde_pedido <= 2:
    estado = "En preparación"
elif dias_desde_pedido <= 5:
    estado = "En camino"
else:
    estado = "Entregado"

print(f"Días: {dias_desde_pedido} → Estado: {estado}")
````

````python
# ¿En qué rango de precio está el producto?
precio = 74.99

if precio >= 100:
    rango = "Premium"
elif precio >= 50:
    rango = "Gama media"
else:
    rango = "Económico"

print(f"Precio: {precio} € → Rango: {rango}")
````

## 📋 Resumen de lo aprendido

| Estructura | Cuándo usarla |
|---|---|
| `if condicion:` | Ejecuta el bloque si la condición es True |
| `elif condicion:` | Alternativa si la anterior no se cumplió |
| `else:` | Caso por defecto si ninguna condición fue True |
