## 🧠 Teoría rápida — if / elif / else

Permite que el programa elija qué hacer según una condición.

```python
ventas = 3450
objetivo = 3000

if ventas >= objetivo:
    print("Objetivo cumplido")
else:
    print("Objetivo no cumplido")
```

**Reglas importantes:**
- Solo se ejecuta el **primer bloque** cuya condición sea verdadera.
- `else` es el caso por defecto si ninguna condición fue True.
- La **indentación** (4 espacios) es obligatoria.

> 💡 En análisis de datos usarás `if/elif/else` dentro de funciones aplicadas con `.apply()` para clasificar o etiquetar valores en una columna.

````python
# ¿Se ha superado el objetivo de ventas del mes?
ventas_mes      = 3450
objetivo_ventas = 3000

if ventas_mes >= objetivo_ventas:
    print("Objetivo cumplido")
else:
    print("Objetivo no cumplido")
````

````python
# ¿El empleado cumple los requisitos para una subida de salario?
# Requisito: más de 2 años en la empresa Y estar activo.
años_empresa = 4
activo       = True

if años_empresa > 2 and activo:
    print("Cumple los requisitos para la subida")
else:
    print("No cumple los requisitos")
````

````python
# ¿El producto tiene stock suficiente para cubrir un pedido?
stock_disponible = 45
unidades_pedido  = 60

if stock_disponible >= unidades_pedido:
    print("Stock suficiente — pedido confirmado")
else:
    print("Stock insuficiente — pedido rechazado")
````

## 🧠 Teoría rápida — elif (múltiples categorías)

Cuando hay más de dos opciones posibles, usamos `elif`:

```python
if ventas >= 5000:
    categoria = "Alto"
elif ventas >= 3000:
    categoria = "Medio"
else:
    categoria = "Bajo"
```

> 💡 Python evalúa las condiciones de arriba a abajo y ejecuta solo la primera que sea True.

````python
# ¿En qué categoría de rendimiento está el empleado según sus ventas?
ventas_mes = 3450

if ventas_mes >= 5000:
    categoria = "Alto rendimiento"
elif ventas_mes >= 3000:
    categoria = "Rendimiento medio"
else:
    categoria = "Bajo rendimiento"

print(f"Ventas: {ventas_mes} € → {categoria}")
````

````python
# ¿En qué tramo salarial está el empleado?
salario = 2850

if salario >= 4000:
    tramo = "Senior"
elif salario >= 2500:
    tramo = "Mid"
else:
    tramo = "Junior"

print(f"Salario: {salario} € → Tramo: {tramo}")
````

````python
# ¿Qué descuento le corresponde al cliente según su volumen de compra?
volumen_compra = 1200

if volumen_compra >= 2000:
    descuento = "20%"
elif volumen_compra >= 1000:
    descuento = "10%"
elif volumen_compra >= 500:
    descuento = "5%"
else:
    descuento = "Sin descuento"

print(f"Compra: {volumen_compra} € → Descuento: {descuento}")
````

---
## 📋 Resumen de lo aprendido

| Estructura | Cuándo usarla |
|---|---|
| `if condicion:` | Ejecuta el bloque si la condición es True |
| `elif condicion:` | Alternativa si la anterior no se cumplió |
| `else:` | Caso por defecto si ninguna condición fue True |
