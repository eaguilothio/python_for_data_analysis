## 🧠 Teoría rápida — Funciones

Una **función** es un bloque de código con nombre que puedes reutilizar.

```python
def calcular_iva(precio):
    iva_cantidad   = precio * 0.21
    precio_con_iva = precio + iva_cantidad
    return round(precio_con_iva, 2)
```

**Llamar a la función:**
```python
calcular_iva(100)   # → 121.0
calcular_iva(49.99) # → 60.49
```

**Parámetros con valor por defecto** — permiten cambiar el IVA si hace falta:
```python
def calcular_iva(precio, iva=0.21):
    iva_cantidad   = precio * iva
    precio_con_iva = precio + iva_cantidad
    return round(precio_con_iva, 2)
```

````python
# Función básica — calcular el precio con IVA
def calcular_iva(precio, iva=0.21):
    iva_cantidad   = precio * iva
    precio_con_iva = precio + iva_cantidad
    return round(precio_con_iva, 2)

print(calcular_iva(100))
print(calcular_iva(49.99))
print(calcular_iva(200, iva=0.10))  # IVA reducido
````

````python
# Función con condición — ¿se ha cumplido el objetivo de ventas?
def evaluar_ventas(ventas, objetivo):
    if ventas >= objetivo:
        resultado = "Objetivo cumplido"
    else:
        resultado = "Objetivo no cumplido"
    return resultado

print(evaluar_ventas(3450, 3000))
print(evaluar_ventas(2800, 3000))
````

````python
# Función con elif — clasificar el rendimiento de un empleado
def clasificar_rendimiento(ventas):
    if ventas >= 5000:
        categoria = "Alto rendimiento"
    elif ventas >= 3000:
        categoria = "Rendimiento medio"
    else:
        categoria = "Bajo rendimiento"
    return categoria

print(clasificar_rendimiento(5200))
print(clasificar_rendimiento(3450))
print(clasificar_rendimiento(1800))
````

````python
# Función aplicada a una lista — clasificar todos los meses del año
ventas = [120, 98, 145, 87, 160, 134, 112, 99, 178, 155, 190, 210]
meses  = ["enero", "febrero", "marzo", "abril", "mayo", "junio",
          "julio", "agosto", "septiembre", "octubre", "noviembre", "diciembre"]

def clasificar_mes(venta):
    if venta >= 160:
        categoria = "Alto"
    elif venta >= 120:
        categoria = "Medio"
    else:
        categoria = "Bajo"
    return categoria

for mes, venta in zip(meses, ventas):
    print(f"{mes}: {venta} k€ → {clasificar_mes(venta)}")
````

````python
# Función con varios parámetros — calcular el margen de beneficio
def calcular_margen(precio_venta, coste):
    beneficio = precio_venta - coste
    margen    = beneficio / precio_venta * 100
    return round(margen, 1)

print(f"Margen: {calcular_margen(149.99, 85)}%")
print(f"Margen: {calcular_margen(99.99, 60)}%")
````

````python
# Función aplicada a una lista de diccionarios
empleados = [
    {"nombre": "Laura García",  "salario": 3100},
    {"nombre": "Carlos Ruiz",   "salario": 2700},
    {"nombre": "Marta Sanz",    "salario": 3400},
    {"nombre": "Iván López",    "salario": 2500},
]

def clasificar_salario(salario):
    if salario >= 3000:
        tramo = "Senior"
    elif salario >= 2500:
        tramo = "Mid"
    else:
        tramo = "Junior"
    return tramo

for empleado in empleados:
    tramo = clasificar_salario(empleado["salario"])
    print(f"{empleado['nombre']}: {empleado['salario']} € → {tramo}")
````

## 📋 Resumen de lo aprendido

| Concepto | Código |
|---|---|
| Crear función | `def nombre(parametro):` |
| Devolver valor | `return valor` |
| Parámetro por defecto | `def f(x, iva=0.21):` |
| Llamar a la función | `nombre(argumento)` |
| Función con condición | `if / elif / else` dentro de `def` |
| Aplicar a una lista | `for x in lista: funcion(x)` |
