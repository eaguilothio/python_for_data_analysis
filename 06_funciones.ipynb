## 🧠 Teoría rápida — Funciones

Una **función** es un bloque de código con nombre que puedes reutilizar.

```python
def calcular_total(precio, cantidad):
    return round(precio * cantidad, 2)
```

**Llamar a la función:**
```python
calcular_total(29.99, 3)   # → 89.97
calcular_total(9.99, 5)    # → 49.95
```

**Parámetros con valor por defecto** — se usan si no se pasa ese argumento:
```python
def calcular_total(precio, cantidad, descuento=0.0):
    total = precio * cantidad * (1 - descuento)
    return round(total, 2)
```

````python
# Función básica — calcular el total de un producto en el carrito
def calcular_total(precio, cantidad, descuento=0.0):
    total = precio * cantidad * (1 - descuento)
    return round(total, 2)

print(calcular_total(29.99, 2))               # sin descuento
print(calcular_total(29.99, 2, descuento=0.10))  # con 10% de descuento
print(calcular_total(89.99, 1, descuento=0.20))  # con 20% de descuento
````

````python
# Función con condición — ¿el pedido tiene envío gratuito?
def evaluar_envio(total_carrito, minimo=40):
    if total_carrito >= minimo:
        return "Envío gratuito"
    else:
        faltan = round(minimo - total_carrito, 2)
        return f"Faltan {faltan} € para envío gratuito"

print(evaluar_envio(53.97))
print(evaluar_envio(22.50))
````

````python
# Función con elif — clasificar un producto por rango de precio
def clasificar_precio(precio):
    if precio >= 100:
        rango = "Premium"
    elif precio >= 50:
        rango = "Gama media"
    else:
        rango = "Económico"
    return rango

print(clasificar_precio(149.99))
print(clasificar_precio(74.99))
print(clasificar_precio(9.99))
````

````python
# Función aplicada a una lista — clasificar todos los productos del catálogo
productos = ["Camiseta", "Pantalón", "Calcetines", "Zapatillas", "Gorra", "Sudadera"]
precios   = [29.99, 49.99, 9.99, 89.99, 15.99, 44.99]

def clasificar_precio(precio):
    if precio >= 60:
        return "Premium"
    elif precio >= 30:
        return "Gama media"
    else:
        return "Económico"

for producto, precio in zip(productos, precios):
    print(f"{producto}: {precio} € → {clasificar_precio(precio)}")
````

````python
# Función con varios parámetros — calcular el total de un carrito completo
def total_carrito(precios, cantidades, descuento=0.0):
    total = sum(p * c for p, c in zip(precios, cantidades))
    total_con_descuento = round(total * (1 - descuento), 2)
    return total_con_descuento

precios    = [29.99, 9.99, 14.99]
cantidades = [1, 3, 2]

print(f"Sin descuento:    {total_carrito(precios, cantidades)} €")
print(f"Con 10% descuento: {total_carrito(precios, cantidades, descuento=0.10)} €")
````

````python
# Función aplicada a una lista de diccionarios
carrito = [
    {"nombre": "Zapatillas running",   "precio": 74.99, "cantidad": 1},
    {"nombre": "Calcetines deportivos", "precio": 9.99,  "cantidad": 3},
    {"nombre": "Botella de agua",       "precio": 14.99, "cantidad": 2},
]

def subtotal_item(item):
    return round(item["precio"] * item["cantidad"], 2)

for item in carrito:
    print(f"{item['nombre']}: {subtotal_item(item)} €")
````

## 📋 Resumen de lo aprendido

| Concepto | Código |
|---|---|
| Crear función | `def nombre(parametro):` |
| Devolver valor | `return valor` |
| Parámetro por defecto | `def f(x, descuento=0.0):` |
| Llamar a la función | `nombre(argumento)` |
| Función con condición | `if / elif / else` dentro de `def` |
| Aplicar a una lista | `for x in lista: funcion(x)` |
