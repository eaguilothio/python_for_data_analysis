## 🧠 Teoría rápida — Bucle for

Un `for` recorre cada elemento de una secuencia uno a uno:

```python
ventas = [120, 98, 145]

for venta in ventas:
    print(venta)
```

Combinado con `if` y `.append()`, permite clasificar valores en listas nuevas:

```python
altas = []
for venta in ventas:
    if venta >= 130:
        altas.append(venta)
```

> ⚠️ En pandas evita iterar filas con `for` — usa operaciones vectorizadas. Usa `for` para iterar sobre listas y diccionarios Python.

````python
# Ventas mensuales del año (en miles de euros)
meses  = ["enero", "febrero", "marzo", "abril", "mayo", "junio",
          "julio", "agosto", "septiembre", "octubre", "noviembre", "diciembre"]
ventas = [120, 98, 145, 87, 160, 134, 112, 99, 178, 155, 190, 210]
````

````python
# ¿Cuánto se vendió cada mes?
for mes, venta in zip(meses, ventas):
    print(f"{mes}: {venta} k€")
````

````python
# ¿Qué meses superaron el objetivo de 130 k€?
objetivo = 130
meses_buenos = []

for mes, venta in zip(meses, ventas):
    if venta >= objetivo:
        meses_buenos.append(mes)

print(f"Meses que superaron el objetivo: {meses_buenos}")
````

````python
# Clasifica cada mes en alto, medio o bajo rendimiento.
alto  = []
medio = []
bajo  = []

for mes, venta in zip(meses, ventas):
    if venta >= 160:
        alto.append(mes)
    elif venta >= 120:
        medio.append(mes)
    else:
        bajo.append(mes)

print(f"Alto:  {alto}")
print(f"Medio: {medio}")
print(f"Bajo:  {bajo}")
````

````python
# Recorre el registro de un empleado e imprime cada campo.
empleado = {
    "nombre":       "Laura García",
    "departamento": "Marketing",
    "salario":      3100,
    "ciudad":       "Barcelona"
}

for campo, valor in empleado.items():
    print(f"{campo}: {valor}")
````

````python
# ¿Qué empleados tienen un salario por encima de 2800 €?
empleados = [
    {"nombre": "Laura García",  "salario": 3100},
    {"nombre": "Carlos Ruiz",   "salario": 2700},
    {"nombre": "Marta Sanz",    "salario": 3400},
    {"nombre": "Iván López",    "salario": 2500},
]

bien_pagados = []

for empleado in empleados:
    if empleado["salario"] > 2800:
        bien_pagados.append(empleado["nombre"])

print(f"Empleados con salario > 2800 €: {bien_pagados}")
````

## 📋 Resumen de lo aprendido

| Concepto | Código |
|---|---|
| Recorrer lista | `for x in lista:` |
| Recorrer dos listas a la vez | `for a, b in zip(lista1, lista2):` |
| Clasificar valores | `if / elif / else` dentro del `for` |
| Añadir a lista | `lista.append(x)` |
| Recorrer diccionario | `for clave, valor in d.items():` |
| Recorrer lista de diccionarios | `for d in lista: d["clave"]` |
