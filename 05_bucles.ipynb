## 🧠 Teoría rápida — Bucle for

Un `for` recorre cada elemento de una secuencia uno a uno:

```python
precios = [29.99, 49.99, 15.99]

for precio in precios:
    print(precio)
```

Combinado con `if` y `.append()`, permite clasificar valores en listas nuevas:

```python
caros = []
for precio in precios:
    if precio >= 40:
        caros.append(precio)
```

> ⚠️ En pandas evita iterar filas con `for` — usa operaciones vectorizadas. Usa `for` para iterar sobre listas y diccionarios Python.

````python
# Productos y precios del catálogo
productos = ["Camiseta", "Pantalón", "Calcetines", "Zapatillas", "Gorra", "Sudadera"]
precios   = [29.99, 49.99, 9.99, 89.99, 15.99, 44.99]
````

````python
# ¿Cuánto cuesta cada producto?
for producto, precio in zip(productos, precios):
    print(f"{producto}: {precio} €")
````

````python
# ¿Qué productos cuestan más de 40 €?
caros = []

for producto, precio in zip(productos, precios):
    if precio > 40:
        caros.append(producto)

print(f"Productos por encima de 40 €: {caros}")
````

````python
# Clasifica cada producto por rango de precio.
economico   = []
gama_media  = []
premium     = []

for producto, precio in zip(productos, precios):
    if precio >= 60:
        premium.append(producto)
    elif precio >= 30:
        gama_media.append(producto)
    else:
        economico.append(producto)

print(f"Premium:    {premium}")
print(f"Gama media: {gama_media}")
print(f"Económico:  {economico}")
````

````python
# Recorre la ficha de un producto e imprime cada campo.
producto = {
    "nombre":    "Zapatillas running",
    "marca":     "Nike",
    "precio":    74.99,
    "stock":     42
}

for campo, valor in producto.items():
    print(f"{campo}: {valor}")
````

````python
# ¿Qué productos del carrito tienen más de 1 unidad?
carrito = [
    {"nombre": "Zapatillas running",   "precio": 74.99, "cantidad": 1},
    {"nombre": "Calcetines deportivos", "precio": 9.99,  "cantidad": 3},
    {"nombre": "Botella de agua",       "precio": 14.99, "cantidad": 2},
]

varios = []

for item in carrito:
    if item["cantidad"] > 1:
        varios.append(item["nombre"])

print(f"Productos con más de 1 unidad: {varios}")
````

## 📋 Resumen de lo aprendido

| Concepto | Código |
|---|---|
| Recorrer lista | `for x in lista:` |
| Recorrer dos listas a la vez | `for a, b in zip(lista1, lista2):` |
| Clasificar valores | `if / elif / else` dentro del `for` |
| Añadir a lista | `lista.append(x)` |
| Recorrer diccionario | `for clave, valor in d.items():` |
| Recorrer lista de diccionarios | `for d in lista: d["clave"]` |
